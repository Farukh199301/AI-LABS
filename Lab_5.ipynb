**QUESTION 1:**

```python
def minimax(depth, nodeIndex, isMax, values, height):
    # Base case: leaf node
    if depth == height:
        return values[nodeIndex]

    if isMax:
        return max(
            minimax(depth + 1, nodeIndex * 2, False, values, height),
            minimax(depth + 1, nodeIndex * 2 + 1, False, values, height)
        )
    else:
        return min(
            minimax(depth + 1, nodeIndex * 2, True, values, height),
            minimax(depth + 1, nodeIndex * 2 + 1, True, values, height)
        )


# Driver code
values = [-4, -1, 2, 3, 4, 5, 6, 7]
height = 3  # tree height

result = minimax(0, 0, True, values, height)
print("Optimal value is:", result)
```

**QUESTION: 2**

```python
import math

def alphabeta(depth, nodeIndex, isMax, values, height, alpha, beta):
    # Leaf node
    if depth == height:
        return values[nodeIndex]

    if isMax:
        best = -math.inf

        for i in range(2):
            val = alphabeta(depth + 1, nodeIndex * 2 + i,
                            False, values, height, alpha, beta)
            best = max(best, val)
            alpha = max(alpha, best)

            # Pruning
            if beta <= alpha:
                break

        return best

    else:
        best = math.inf

        for i in range(2):
            val = alphabeta(depth + 1, nodeIndex * 2 + i,
                            True, values, height, alpha, beta)
            best = min(best, val)
            beta = min(beta, best)

            # Pruning
            if beta <= alpha:
                break

        return best


# Driver code
values = [4, 3, 8, 7, 0, 1, 2, 5]
height = 3

result = alphabeta(0, 0, True, values, height, -math.inf, math.inf)
print("Optimal value is:", result)
```
